```python
#library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("insta.csv",encoding="latin_1")
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Impressions From Home")
sns.kdeplot(df['From Other'],fill=True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.style.use("fivethirtyeight")

sns.distplot(df["From Other"])

plt.show()
```

```python
#line plot
plt.figure(figsize=(10,8))
plt.style.use("fivethirtyeight")
sns.lineplot(df['From Home'])

plt.xlim(0)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.style.use("fivethirtyeight")
pallete=sns.color_palette("pastel")
sns.barplot(pallete)
plt.xlim(100)

sns.barplot(x=df.index, y=df["From Home"], palette="muted")
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(df['Comments'])
plt.style.use("fivethirtyeight")

plt.show()
```

```python
#this the way to plot pie .
home=df['From Home'].sum()
hashtag=df['From Hashtags'].sum()
explore=df['From Explore'].sum()
other=df['From Other'].sum()
values=[home,hashtag,explore,other]
labels=['from home',"from hashtag","from explore","from other"]
fig= px.pie(df,values=values,names=labels,hole=0.5)
fig.show()
```

```python
#scatter plot using plotly express
fig=px.scatter(data_frame=df,x='Impressions',y='Likes',trendline="ols",size='Likes')
fig.show()
```
