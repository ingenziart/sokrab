```python
training_set = x_train.iloc[:,4].values
#testing_set=pd.DataFrame(training_set,columns=['close'])
#testing_set.tail()
training_set
```

```python
testing_set=x_test.iloc[:,4].values
#testing_set=pd.DataFrame(x_test,columns=["close"])
#testing_set.tail()
testing_set
```

```python
training_set = training_set.reshape(-1, 1)
testing_set = testing_set.reshape(-1, 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
train_scarer=scaler.fit_transform(training_set)
test_scarer=scaler.transform(testing_set)
```

```python

#create lstm

def create_sequence(train_scarer,seq_length=60):
  X_train=[]
  Y_train=[]
  for i in range(seq_length,len(train_scarer)):
    X_train.append(train_scarer[i-seq_length:i,0])
    Y_train.append(train_scarer[i,0])

  return X_train,Y_train

X_train, y_train = create_sequence(train_scarer, seq_length=60)
X_test, y_test   = create_sequence(test_scarer, seq_length=60)


```
